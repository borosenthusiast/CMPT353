```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
data = [None] * 6
data[0] = pd.read_csv('data-1.csv')
data[1] = pd.read_csv('data-2.csv')
data[2] = pd.read_csv('data-3.csv')
data[3] = pd.read_csv('data-4.csv')
data[4] = pd.read_csv('data-5.csv')
data[5] = pd.read_csv('data-6.csv')
```

```python
def generate_summary(data, idx):
    #Mean
    dmean = data.mean(axis=0)
    #print("Mean")
    #print(dmean, end='\n\n')
    #print(dmean.x)
    #Standard Deviation
    dstd = data.std(axis=0)
    #print("Standard Deviation")
    #print(dstd, end='\n\n')

    dmin = data.min(axis=0)
    #print("Min")
    #print(dmin, end='\n\n')

    dmax = data.max(axis=0)
   # print("Max")
    #print(dmax, end='\n\n')

    ranged = dmax - dmin
    #print("Range")
    #print(ranged, end='\n\n')

    dcorr = data.x.corr(data.y, method='pearson')
   # print('Correlation Coefficient')
    #print(dcorr)
    
    df =  pd.DataFrame(
        {
            "Mean X": dmean.x,
            "Mean Y": dmean.y,
            "Stdev X": dstd.x,
            "Stdev Y": dstd.y,
            "Min X": dmin.x,
            "Min Y": dmin.y,
            "Max X": dmax.x,
            "Max Y": dmax.y,
            "Range X": ranged.x,
            "Range Y": ranged.y,
            "Correlation Coefficient": dcorr
        },
            index = ["Dataset" + str(idx)]
        )
    #print(df)
    return df
```

```python
summary = pd.DataFrame()
i = 1
for d in data:
    summary = summary.append(generate_summary(d, i))
    i = i + 1
summary  
```

```python
with open('summary.txt','w') as output:
    summary.to_string(output)
```
